```python
import matplotlib.pyplot as plt
import numpy as np

# bir data oluşturalım
# -10 ile 9 arasında 20 eşit uzaklıkta bulunan sayı üretsin
x = np.linspace(-10,9,20)
y = x**3

figure = plt.figure()

axes = figure.add_axes([0.2,0.2,0.8,0.8])  # 0.2 soldan grafiği sağa doğru %20 getirecek anlamında
                                    # diğer 0.2 aşşağıdan ne kadar geleceği
                                    # 0.8 genişlik ne kadar olsun
                                    # diğer 0.8 yükseklik ne kadar olsun

plt.show()  # figür üzerinde boş bir axes geldi bize
```

```python
import matplotlib.pyplot as plt
import numpy as np

# bir data oluşturalım
x = np.linspace(-10,9,20)
y = x**3

figure = plt.figure()

axes = figure.add_axes([0.1,0.1,0.8,0.8])  

# çizgimizi hazırlayalım
axes.plot(x,y,"blue")

# x ve y eksenine yine isim verelim
axes.set_xlabel("X Axis")
axes.set_ylabel("X Axis")
axes.set_title("Cube")


plt.show()
```

* figüre bir axes daha ekleyelim

```python
import matplotlib.pyplot as plt
import numpy as np

# bir data oluşturalım
x = np.linspace(-10,9,20)
y = x**3
z = x**2    # bir axes daha ekledik

figure = plt.figure()

axes_1 = figure.add_axes([0.1,0.1,0.8,0.8])     # ismini axes_1 yaptık


axes_1.plot(x,y,"blue")

axes_1.set_xlabel("X Axis")
axes_1.set_ylabel("X Axis")
axes_1.set_title("Cube")

axes_2 = figure.add_axes([0.15,0.6,0.25,0.25])

# yeni çizgimiz
axes_2.plot(x,z,"red")

# yeni eklenen x ve y eksenlerinin isimleri
axes_2.set_xlabel("X Axis")
axes_2.set_ylabel("X Axis")
axes_2.set_title("Square")

plt.show()
```
