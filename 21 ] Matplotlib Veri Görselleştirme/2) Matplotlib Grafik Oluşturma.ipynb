```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,100)

plt.plot(x, x, label= "linear") # label parametresi ile isimlerini belirtmiş olduk sadece
plt.plot(x, x**2, label = "quadratic")
plt.plot(x, x**3, label= "cubic")

# sol ve alt çizginin isimlerini belirtiyor öğrenmiştik
plt.xlabel("x label")
plt.ylabel("y label")

plt.title("simple plot")    # başlık verelim grafiğe

plt.legend()    # çizgilerin isimlerini yazdıysak bu da sadece tabloda belirtiyor
plt.show()
```

Bir düzleme birden fazla grafik koyalım

        axs()
        subplots()

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,100)
# fig'i sonraki ders işleyeceğiz
fig,axs = plt.subplots(3)   # 2 verince 2 tane axs alanı oluşturmuş oluyoruz
                            # kordinat sisteminin her bir köşesi & bölgesi bir aks .
axs[0].plot(x, x, color = "red")
axs[1].plot(x, x**2, color = "green")
axs[2].plot(x, x**3, color = "yellow")

plt.show()
```

* devam

        set_title("linear")

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,100)

fig,axs = plt.subplots(3)   
                            
axs[0].plot(x, x, color = "red")
axs[0].set_title("linear")  # isim ekledik
axs[1].plot(x, x**2, color = "green")
axs[1].set_title("quadratic")   # isim ekledik
axs[2].plot(x, x**3, color = "yellow")
axs[2].set_title("cubic")   # isim ekledik

plt.show()
```

başlıklar birbirine girdi düzeltmek için

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,100)

fig,axs = plt.subplots(3)   
                            
axs[0].plot(x, x, color = "red")
axs[0].set_title("linear") 
axs[1].plot(x, x**2, color = "green")
axs[1].set_title("quadratic")   
axs[2].plot(x, x**3, color = "yellow")
axs[2].set_title("cubic")   

plt.tight_layout()  # eklendi 
plt.show()
```

* Alt alta grafik oluşturmak yerine 4 bölmeli bir grafik oluşturalım

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,100)

fig,axs = plt.subplots(2,2)  # iki satır iki kolon anlamında  
# figüre başlık verelim
fig.suptitle("grafik başlığı")

axs[0,0].plot(x, x, color = "red")  
axs[0,1].plot(x, x**2, color = "blue")  # sıfırıncı satır 1. kolon anlamında
axs[1,0].plot(x, x**3, color = "green") # birinci satır 0. kolon
axs[1,1].plot(x, x**4, color = "yellow")
                            
plt.show()
```

* bir dataFrame kullanalım veri olarak

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("nba.csv")

# dataFrame deki Number isimli kolonu silelim ve takım bilgisine göre gruplayalım ve geri kalan kolonlar için ortalama aldıralım önce
df = df.drop(["Number"], axis= 1).groupby("Team").mean()   # axis 1 kolon sileceğimiz için
df.plot(subplots = True)    # True dediğimiz için her biri farklı bir grafiğe yerleştirilmiş oldu

plt.legend()
plt.show()
```

* yukarıdakinin sadece dataframe deki ilk 5 bilgiyi alacak şekilde düzenleyelim

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("nba.csv")

# dataFrame deki Number isimli kolonu silelim ve takım bilgisine göre gruplayalım ve geri kalan kolonlar için ortalama aldıralım önce
df = df.drop(["Number"], axis= 1).groupby("Team").mean()   # axis 1 kolon sileceğimiz için
df.head(5).plot(subplots = True)    # True dediğimiz için her biri farklı bir grafiğe yerleştirilmiş oldu

plt.legend()
plt.show()
```
