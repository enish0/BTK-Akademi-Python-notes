* json kullanarak stringi dict e d√∂n√º≈üt√ºrelim

```python
import requests
import json # json string den dict e d√∂n√º≈üt√ºrmek i√ßin

result = requests.get("https://jsonplaceholder.typicode.com/todos")

print("üêõdict halde sitedeki t√ºm bilgilerüêõ".center(50," "))
result = json.loads(result.text)    # d√ºzenlendi
print(result)   # sitedeki t√ºm bilgiyi busefer json stringinden dict e √ßevirilmi≈ü halde getirdik buraya

# artƒ±k i√ßinden cƒ±mbƒ±zla bilgi alabilriz (dict halde)
print("üêõi√ßinden teker teker ula≈ütƒ±klarƒ±mƒ±züêõ".center(50," "))
print(result[0])
print(result[0]["title"])
```

* for kullanarak gelen bilgileri satƒ±r satƒ±r yazƒ±lsƒ±n d√ºzenli

```python
import requests
import json 

result = requests.get("https://jsonplaceholder.typicode.com/todos")

print("üêõdict halde sitedeki t√ºm bilgilerüêõ".center(50," "))
result = json.loads(result.text)   
print(result)   

# ------yukarƒ±sƒ± deƒüi≈ümedi--------

# bu sefer daha d√ºzenli bir ≈üekilde alalƒ±m

for i in result:
    print(i)
```

* peki bir filtreleme yapmak istersem ?

```python
import requests
import json 

result = requests.get("https://jsonplaceholder.typicode.com/todos")

print("üêõdict halde sitedeki t√ºm bilgilerüêõ".center(80,"-"))
result = json.loads(result.text)   
print(result)

# ------yukarƒ±sƒ± deƒüi≈ümedi--------
print("üêõsadece userId'si 1 olanlarƒ±n title ƒ±nƒ±üêõ".center(80,"-"))

for i in result:
    if i["userId"] == 1: # sadece userId'si 1 olanlarƒ±n title ƒ±nƒ± yaz demek istedik
        print(i["title"])
```
