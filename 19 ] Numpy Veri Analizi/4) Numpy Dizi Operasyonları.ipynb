# Numpy Dizi Operasyonları

```python
import numpy as np

# rastgele iki farklı dizi oluşturacak
numbers1 = np.random.randint(10,100,6)  # 10 ile 100 arası rastgele sayı üretsin
numbers2 = np.random.randint(10,100,6)

print(numbers1)
print(numbers2)

# rastgele oluşturduğumuz diziyi toplayabiliriz.

result = numbers1 + numbers2
print(result)   # iki dizinin aynı indexteki elemanlarını topluyor ve dizi olarak grupluyor.

# veya şöyle de yapabiliriz
result = numbers1 + 10
print(result)   # tüm indextekilere 10 eklemiş olduk

# farklarını da alabiliriz
result = numbers1 - numbers2
print(result)

# şöyle yapalım
result = numbers1 - 10
print(result)   # tüm elemanlardan 10 çıkarılır

# çarpma işlemi de olur
result = numbers1 * numbers2
print(result)

# bölme işlemi de olur
result = numbers1 / numbers2
print(result)


# herhangi bir dizinin sinüsünü alalım
result = np.sin(numbers1)
print(result)

# cosinus ünü alalım
result = np.cos(numbers1)
print(result)

# sqrt metoduna verirsek
result = np.sqrt(numbers1)  # her bir elemanını karekökünü alır
print(result)

# logaritmasını alalım
result = np.log(numbers1)
print(result)


# multinumbers yapıp ([] ile gruplara ayırıp) işlemlere öyle devam edelim
mnumbers1 = numbers1.reshape(2,3)
mnumbers2 = numbers2.reshape(2,3)   # 2 satır 3 de kolon(sütun) olarak ayırdık(grupladık)

print(mnumbers1)
print(mnumbers2)

# bu gelen matrisleri birleştirmek istersek eğer...
result = np.vstack((mnumbers1,mnumbers2))   # vstack() dikey olarak birleştirir matrisleri
print("sa")
print(result)

result = np.hstack((mnumbers1, mnumbers2))  # hstack() yatay olarak birleştirdi matrixi
print(result)

# elemanların belli bir koşula uyup uymadığını kontrol edelim

# elemanların her biri için 5 ten büyük mü..
result = numbers1 >= 5
print(result)  # bool türünde değer gönderir

# şimdi 50 den büyük mü onu yapalım
result = numbers1 >= 50
print(result)

# (sayı çift mi tek mi) numbers1 den gelen her bir değerin mod iki sinin sıfıra eşit olup olmadığını kontrol edelim
result = numbers1 %2 == 0
print(result)

# True olan sonuçlar ayrı bir liste içersinde bize geri dönsün istersek
print(numbers1[result])
```
