        10- Kategoriye g√∂re beƒüeni ortalamalarƒ±nƒ± sƒ±ralƒ± ≈üekilde getiriniz.

        13- Her videonun title uzunluƒüu bilgisini yeni bir kolonda g√∂steriniz.

```python
import pandas as pd
df = pd.read_csv("youtube-ing.csv")
#--- temel kod

df["title_len"] = df["title"].apply(len)

result = df
print(result.head(3))
```

        14- Her video i√ßin kullanƒ±lan tag sayƒ±sƒ±nƒ± yeni kolonda g√∂steriniz.

```python
import pandas as pd
df = pd.read_csv("youtube-ing.csv")
#--- temel kod

# taglarƒ± split ile | dan ayrƒ±yoruz liste haline gelen yeni taglarƒ± da len ile sayƒ±sƒ±nƒ± hesaplamƒ±≈ü oluyoruz
df["tag_count"] = df["tags"].apply(lambda x: len(x.split('|')))

result = df
print(result.head(1))
```

* alternatif olarak

```python
import pandas as pd
df = pd.read_csv("youtube-ing.csv")
#--- temel kod

# bu sefer lambda ile deƒüil de normal fonksiyon gibi yapalƒ±m

def tagCount(tag):
    return len(tag.split('|'))

df["tag_count"] = df["tags"].apply(tagCount)
result = df
print(result.head(1))
```

        15- En pop√ºler videolarƒ± listeleyiniz.(like/dislike oranƒ±na g√∂re)

```python
import pandas as pd
df = pd.read_csv("youtube-ing.csv")
#--- temel kod

# dataFrame deki her bir videonun like & dislike sayƒ±sƒ±nƒ± bir liste i√ßine alalƒ±m.
def likeDislikeOranHesapla(dataset):
    likesList = list(df["likes"])   # list metodu otomatik liste i√ßine aldƒ±
    dislikesList = list(df["dislikes"])

    # bize iki bilgi tuple listesi olarak gelsin
    liste = list(zip(likesList, dislikesList))   # zip metodu tuple listesi yapƒ±yor # birde list ile liste ye de √ßevirdik
    print(liste)    # liste i√ßinde tuple lar var

    # liste halindeki like dislike oranlarƒ± hesaplamak istiyoruz
    oranListesi = []
    for like,dislike in liste:
        if (like + dislike) == 0:   # like ve dislike bilgisi 0 ise
            oranListesi.append(0)   # beƒüenilme oranƒ± 0 olarak oranListesine eklensin
        else:   # diƒüer durum yani like & dislike bilgisi 0 deƒüilse hesaplanacak oran vardƒ±r
            oranListesi.append(like/(like+dislike))
    return oranListesi

# hatta bir kolon ekleyip bu oranlarƒ± orada g√∂sterelim
df["beƒüeni_orani"] = likeDislikeOranHesapla(df)

print(df.sort_values("beƒüeni_orani", ascending= False).head()[["title","likes","dislikes","beƒüeni_orani"]])
```
